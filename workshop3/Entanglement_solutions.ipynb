```python
# import statements
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
```

#### _1. Exercise_

a. $|\psi\rangle = \frac{1}{\sqrt{3}} |0\rangle + \sqrt{\frac{2}{3}} |1\rangle$

$$\alpha = |\frac{1}{\sqrt{3}}|² = \frac{1}{3}$$
$$\beta= |\sqrt{\frac{2}{3}}|² = \frac{2}{3}$$

b. $|\psi\rangle = i\frac{2}{3} |0\rangle + \frac{\sqrt{5}}{3} |1\rangle$

$$\alpha = |i\frac{2}{3}|² = \frac{4}{9}$$
$$\beta= |\frac{\sqrt{5}}{3} |² = \frac{5}{9}$$

#### _3. Exercise_

```python
# Phi-Minus
phi_minus = QuantumCircuit(2)

phi_minus.x(0)
phi_minus.h(0)
phi_minus.cx(0,1)

phi_minus.measure_all()
phi_minus.draw(output='mpl')
```

```python
simulator = AerSimulator()
plot_histogram(simulator.run(phi_minus).result().get_counts())
```

```python
# Psi-Plus
psi_plus = QuantumCircuit(2)

psi_plus.h(0)
psi_plus.x(1)
psi_plus.cx(0,1)

psi_plus.measure_all()
psi_plus.draw(output='mpl')
```

#### 4a. Alice sends the $|01\rangle$ state
$|00\rangle \xrightarrow{H} \frac{|00\rangle + |10\rangle}{\sqrt{2}} \xrightarrow{CNOT} \frac{|00\rangle + |11\rangle}{\sqrt{2}} \xrightarrow{X} \frac{|10\rangle + |01\rangle}{\sqrt{2}} \xrightarrow{CNOT} \frac{|11\rangle + |01\rangle}{\sqrt{2}} = |+\rangle \otimes |1\rangle \xrightarrow{H} |01\rangle$ 

#### 4b. In order to send $|10\rangle$ Alice has to use the Z gate
$|00\rangle \xrightarrow{H} \frac{|00\rangle + |10\rangle}{\sqrt{2}} \xrightarrow{CNOT} \frac{|00\rangle + |11\rangle}{\sqrt{2}} \xrightarrow{Z} \frac{|00\rangle - |11\rangle}{\sqrt{2}} \xrightarrow{CNOT} \frac{|10\rangle - |01\rangle}{\sqrt{2}} = |-\rangle \otimes |+\rangle \xrightarrow{H} |10\rangle$
