# The Bloch Sphere

### What is the Bloch Sphere?

- a representation of the state of a qubit

### Implementation in Qiskit

```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector, Operator
from qiskit.visualization import plot_bloch_multivector, plot_bloch_vector
```

```python
qc = QuantumCircuit(1)
qc.x(0)
```

####  Bloch visualisation of a two-qubit quantum circuit:

```python
state = Statevector(qc)
plot_bloch_multivector(state, title= "QUTE 2025")
```

```python
qc = QuantumCircuit(2)

qc.h(0)
qc.x(0)
qc.h(0)
qc.z(1) 

qc.draw(output='mpl')
```

```python
state = Statevector(qc)
plot_bloch_multivector(state, title= "QUTE 2025")
```

You can also use cartesian coordinates:

```python
plot_bloch_vector([0,0,1], title="QUTE 2025")
```

Or using spherical coordinates:

```python
import numpy as np
```

```python
plot_bloch_vector([1, np.pi/2, np.pi/4], coord_type='spherical')
```
