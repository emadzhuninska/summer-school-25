```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector, Operator
from qiskit.visualization import plot_bloch_multivector, plot_bloch_vector
```

```python
qc_a = QuantumCircuit(1)

qc_a.x(0)
qc_a.h(0)
plot_bloch_multivector(Statevector(qc_a), title= "Exercise 2a")
```

```python
qc_b = QuantumCircuit(1)

qc_b.x(0)
qc_b.h(0)
qc_b.z(0)

print(Operator(qc_b))
print(Statevector(qc_b))

plot_bloch_multivector(Statevector(qc_b), title= "Exercise 2b")
```

```python
import numpy as np
plot_bloch_vector([1, np.pi/2, np.pi/2], coord_type='spherical', title="Exercise 2c")
```
