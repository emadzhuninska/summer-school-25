## Solutions for Circuit Composition

```python
import matplotlib.pyplot as plt
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram

# Suppress warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
def get_counts(circuit):
    backend = AerSimulator()
    return backend.run(circuit).result().get_counts()
```

#### _Exercise 1_
Calculate the matrices for $|\psi\rangle$ and sketch the circuits:

a. $XH|0\rangle$

$$H|0\rangle = \frac{1}{\sqrt{2}} \begin{bmatrix} 1&1\cr1&-1\end{bmatrix} \begin{bmatrix}1 \cr 0\end{bmatrix} =\frac{1}{\sqrt{2}} \begin{bmatrix}1 \cr 1\end{bmatrix}$$

$$X|+\rangle = \begin{bmatrix} 0&1\cr1&0\end{bmatrix} \frac{1}{\sqrt{2}}\begin{bmatrix}1 \cr 1\end{bmatrix} = \frac{1}{\sqrt{2}}\begin{bmatrix}1 \cr 1\end{bmatrix}$$

b. $ZX|0\rangle$

$$X|0\rangle =  \begin{bmatrix} 0&1\cr1&0\end{bmatrix} \begin{bmatrix}1 \cr 0\end{bmatrix} = \begin{bmatrix}0 \cr 1\end{bmatrix}$$

$$Z|1\rangle = \begin{bmatrix} 1&0\cr0&-1\end{bmatrix} \begin{bmatrix}0 \cr 1\end{bmatrix} = -\begin{bmatrix}0 \cr 1\end{bmatrix}$$
c. $HXH|1\rangle$

$$H|1\rangle = \frac{1}{\sqrt{2}} \begin{bmatrix} 1&1\cr1&-1\end{bmatrix} \begin{bmatrix}0 \cr 1\end{bmatrix} = \frac{1}{\sqrt{2}} \begin{bmatrix}1 \cr -1\end{bmatrix}$$
$$X|-\rangle = \begin{bmatrix} 0&1\cr1&0\end{bmatrix} \frac{1}{\sqrt{2}}\begin{bmatrix}1 \cr -1\end{bmatrix} = \frac{1}{\sqrt{2}}\begin{bmatrix}-1 \cr 1\end{bmatrix}$$
$$H-|-\rangle = \frac{1}{\sqrt{2}} \begin{bmatrix} 1&1\cr1&-1\end{bmatrix} \frac{1}{\sqrt{2}} \begin{bmatrix}-1 \cr 1\end{bmatrix} = \frac{1}{2} \begin{bmatrix}0 \cr -2\end{bmatrix} = - \begin{bmatrix}0 \cr 1\end{bmatrix}$$

#### _Exercise 2_

```python
qc_a = QuantumCircuit(2)

qc_a.x(1)
qc_a.cx(1,0)
qc_a.x(1)

qc_a.measure_all()
qc_a.draw(output='mpl')
```

```python
plot_histogram(get_counts(qc_a))
```

```python
qc_b = QuantumCircuit(2)

qc_b.x(0)
qc_b.z(1)
qc_b.cx(0,1)
qc_b.z(0)

qc_b.measure_all()
qc_b.draw(output="mpl")
```

```python
plot_histogram(get_counts(qc_b))
```

```python
qc_c = QuantumCircuit(2)

qc_c.h(0)
qc_c.z(0)
qc_c.x(0)
qc_c.h(1)
qc_c.x(1)
qc_c.h(1)
qc_c.measure_all()

qc_c.draw(output='mpl')
```

```python
plot_histogram(get_counts(qc_c))
```

```python
large = QuantumCircuit(5, 3)

large.h(0)
large.h(1)
large.cx(0,2)
large.cx(1,2)
large.t(1)
large.z(3)
large.barrier()
large.cx(2,4)
large.x(2)
large.swap(1,3)
large.barrier()
large.measure(0,0)
large.measure(2,1)
large.measure(4,2)

large.draw(output='mpl')
```
